```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the Excel file
data=pd.read_csv('HR Analytics Data.csv')
df=pd.DataFrame(data)
print(df)
# Display basic information about the data
#print("Data Overview:")
print(df.head())
#print("\nData Info:")
print(df.info())
#("\nData Description:")
print(df.describe())

# Save a summary of the data to a CSV file
df.describe().to_csv('data_summary.csv')
#Plot the distribution of a numerical column
sns.histplot(df['Age'], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.savefig('age_distribution.png')
plt.show()
#PLOT BUSINESS TRAVEL BY DEPARTMENT
sns.histplot(data=df, x='BusinessTravel', hue='Department', multiple='stack', kde=True)
plt.title('Business Travel Distribution by Department')
plt.xlabel('Business Travel Frequency')
plt.ylabel('Count')
plt.savefig('Business_travel_by_department.png')
plt.show()

ax = sns.countplot(data=df, x='Gender', hue='Department')

# Add data labels
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', 
                (p.get_x() + p.get_width() / 2., height),
                ha = 'center', va = 'baseline', fontsize=10, color='black', xytext=(0, 5),
                textcoords = 'offset points')

plt.title('Gender by Department')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.savefig('business_travel_department_barplot_with_labels.png')
plt.show()
print(df.columns)

# Create a bar plot with data labels
ax = sns.countplot(data=df, x='Department', hue='JobRole')

# Add data labels
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', 
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='baseline', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.title('Job Role in each department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.savefig('business_travel_jobrole_barplot_with_labels.png')
plt.show()


```
